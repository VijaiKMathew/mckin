```python
categoricial_feature_names = ['sourcing_channel','residence_area_type']
numeric_feature_names=['perc_premium_paid_by_cash_credit',  'Income','application_underwriting_score',
       'no_of_premiums_paid','age_in_days','premium'] #'age_in_days',
hour_df=pd.get_dummies(hour_df, columns=categoricial_feature_names)
```

```python
from sklearn.preprocessing import MinMaxScaler
ss = MinMaxScaler()
# fit scaler on numeric features
ss.fit(hour_df[numeric_feature_names])

# scale numeric features now
hour_df[numeric_feature_names] = ss.transform(hour_df[numeric_feature_names])

# view updated featureset
```
