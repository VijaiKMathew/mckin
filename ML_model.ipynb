```python
# get features and corresponding outcomes
feature_names = ['perc_premium_paid_by_cash_credit', 'age_in_days', 'Income',
       'Count_3-6_months_late', 'Count_6-12_months_late',
       'Count_more_than_12_months_late', 'application_underwriting_score',
       'no_of_premiums_paid', 'premium', 'sourcing_channel_A',
       'sourcing_channel_B', 'sourcing_channel_C', 'sourcing_channel_D',
       'sourcing_channel_E', 'residence_area_type_Rural',
       'residence_area_type_Urban']
training_features = df[feature_names]

outcome_name = ['renewal']
outcome_labels = df[outcome_name]
```

```python
from sklearn.linear_model import LogisticRegression
import numpy as np

# fit the model
lr = LogisticRegression() 
model = lr.fit(training_features, np.array(outcome_labels['renewal']))
# view model parameters
model
```

```python
def print_score(clf, X_train, y_train, X_test, y_test, train=True):
    if train:
        print("Train Result:\n")
        print("accuracy score: {0:.4f}\n".format(accuracy_score(y_train, clf.predict(X_train))))
        print("Classification Report: \n {}\n".format(classification_report(y_train, clf.predict(X_train))))
        print("Confusion Matrix: \n {}\n".format(confusion_matrix(y_train, clf.predict(X_train))))

        res = cross_val_score(clf, X_train, y_train, cv=10, scoring='accuracy')
        print("Average Accuracy: \t {0:.4f}".format(np.mean(res)))
        print("Accuracy SD: \t\t {0:.4f}".format(np.std(res)))
        print("train auc_roc",roc_auc_score(y_train, clf.predict(X_train)))
    elif train==False:
        print("Test Result:\n")        
        print("accuracy score: {0:.4f}\n".format(accuracy_score(y_test, clf.predict(X_test))))
        print("Classification Report: \n {}\n".format(classification_report(y_test, clf.predict(X_test))))
        print("Confusion Matrix: \n {}\n".format(confusion_matrix(y_test, clf.predict(X_test)))) 
        print("test auc_roc",roc_auc_score(y_test, clf.predict(X_test)))
```

```python
from sklearn.model_selection import cross_val_predict, cross_val_score
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix,roc_curve,roc_auc_score

# evaluate model performance

```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(training_features, np.array(outcome_labels['renewal']),
                                                    train_size=0.7,
                                                    random_state=42)
```

```python
lr = LogisticRegression()  
model = lr.fit(X_train, y_train)
```

```python
print_score(model, X_train, y_train, X_test, y_test, train=True)
```

```python
print_score(model, X_train, y_train, X_test, y_test, train=False)
```
